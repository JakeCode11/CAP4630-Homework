```
## Jacob Belga
## CAP 4630 - Introduction to AI
## HW 1 - Problem 1

from mpl_toolkits import mplot3d
from scipy.signal import argrelextrema
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def f1(x, y):
  return x**2 + y**2

def f2(x, y):
  return -x**2 - y**2

def f3(x, y):
  return x**2 - y**2

x = np.linspace(-6, 6, 80)
y = np.linspace(-6, 6, 80)

X, Y = np.meshgrid(x,y)

## Local Minimum Function call
Z1 = f1(X,Y)
z1 = f1(x,y)
## Local Maximum Function call
Z2 = f2(X,Y)
z2 = f2(x,y)
## Zero Gradient Function call
Z3 = f3(X,Y)
z3 = f3(x,y)

## Plotting local min graph
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z1, 50, cmap ='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x, y, z1, c=z1, cmap='Reds', linewidth=1);
ax.view_init(20, 25)

## Plotting local max graph
fig2 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z2, 50, cmap ='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter3D(x, y, z2, c=z2, cmap='Blues', linewidth=1);
ax.view_init(20, 25)

## Plotting neither max or min graph
fig3 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z3, 50, cmap ='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(x, y, z3, c=z3, cmap='Reds', linewidth=1);
ax.view_init(20, 50)



   
```
