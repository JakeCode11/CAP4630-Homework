```
## Jacob Belga
## CAP 4630 - Introduction to AI
## HW1 - Problem 2

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
%matplotlib inline

fig1 = plt.figure()
ax = plt.axes(projection='3d')

np.random.seed(42)

xs = 2 * np.random.rand(100, 1)
ys = 4 + 3 * xs + np.random.rand(100, 1)
zs = ys

ax.scatter3D(xs, ys)
```

```
## Splitting the data into training and test sets

## Training Set 
train_xs = xs[:80]
train_ys = ys[:80]
train_zs = zs[:80]

## Testing Set
test_xs = xs[80:]
test_ys = ys[80:]
test_zs = zs[80:]
```

```
## Number of Epochs
epochs = 10
## Learning Rate
lr = 0.01
```

```
## Initial values for weights w1 and w2 and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_xs[i] + b
    y_pred2 = w2 * train_zs[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_xs[i]
    grad_w2 = (y_pred2 - train_ys[i]) * train_zs[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0

for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_xs[i] + w2 * test_zs[i] + b - test_ys[i]) ** 2
test_loss /= 20
```

```
pred_ys = w1 * test_xs + w2 * test_zs + b
```

```
fig2 = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(test_xs, test_ys, test_zs)
ax.scatter3D(test_xs, pred_ys, test_zs)
```
